```python
r = requests.get('http://api.clubelo.com/Liverpool')
```

```python
r.status_code
```

```python
r.text
```

```python
data = StringIO(r.text)
df = pd.read_csv(data, sep=",")

df.tail()
```

```python
df.From = pd.to_datetime(df['From'])
df.To = pd.to_datetime(df['To'])
```

```python
sns.set_style("dark")

fig=plt.figure (num=None, figsize=(15,5), dpi = 100)
axes = fig.add_subplot(1,1,1, facecolor='black')
fig.patch.set_facecolor('black')
axes.tick_params(axis='x', colors = 'white')
axes.tick_params(axis='y', colors = 'white')

plt.plot_date(df.From, df.Elo, '-', color = "red")

#Set a title, write it on the left hand side
plt.title("Liverpool - Elo Rating", loc="left", fontsize = 15, color = 'white')
plt.xlabel('Year', color = 'white')
plt.ylabel('Elo', color = 'white')

plt.show()
```
