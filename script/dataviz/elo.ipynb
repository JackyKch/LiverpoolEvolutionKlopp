## Imports

```python
import pandas as pd 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.patheffects as path_effects
import datetime
```

```python
data = pd.read_csv("../../data/all_season/Liverpool_data_epl_match_details.csv")
```

```python
data_ldc = pd.read_csv("../../data/all_season/Liverpool_data_ldc_match_details.csv")
```

```python
data_ldc.rename(columns={"elo_liverpool": "Elo_Liverpool"}, inplace=True)
```

```python
data = data.append(data_ldc, ignore_index=True)
```

```python
data = data.sort_values(by=['date_match'])
```

```python
elo_liverpool = data['Elo_Liverpool'] 
date_match = data['date_match']
```

```python
data.loc[data['date_match'] == "2015-10-17", "Elo_Liverpool"]
```

```python
fig,ax = plt.subplots(figsize = (11,7))

#Add grid to the plot
ax.grid(ls="dotted", lw="0.5", color="w", zorder=1)

hfont = {'fontname':'Impact'}

#Change the background color
fig.set_facecolor("#08143D")
ax.patch.set_facecolor("#08143D")

#Change the ticks color
ax.tick_params(colors="w", length=15)

#Change the axes color
ax.spines['bottom'].set_color("w")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color("w")

# Plot the elo and the date 
plt.plot(date_match, elo_liverpool, zorder=2, color="#2bb6e0" )

# Set the ticks
plt.xticks(['2012-08-18', '2014-04-27', '2015-10-17','2017-08-12', '2018-08-12', "2019-06-01", '2020-03-07'])
plt.yticks([1700, 1800, 1900, 2000, 2100])
ax.set_ylim(1700)
plt.xlim(xmin='2012-08-18')

# Axes Label
plt.xlabel("Date", color="w", fontsize=15 , **hfont)
plt.ylabel("Elo", color="w", fontsize=15 , **hfont)

# Title and credits
fig.text(0.12,0.9, "Liverpool Elo ratings since 2012/2013 season\n", fontsize = 16, fontweight = "bold", color = "w", **hfont)
fig.text(0.12,0.9, "Comparing Rodgers and Klopp teams ratings in Premier League and Champions League", fontsize = 12, fontweight="regular", color = "w", **hfont)
fig.text(0.05, 0, "Created by Jacky KUOCH",
        fontstyle="italic",fontsize=9, color="w")

# Split the graph in two
ax.axvline(x="2015-10-17", color="w")
rodgerspart = ax.text(40, 2050, "Rodgers", fontsize=18, ha="center", color="w",alpha=0.2 , **hfont)
klopppart = ax.text(160, 2050, "Klopp", fontsize=18, ha="center", color="w",alpha=0.2 , **hfont)
# Add points on key dates

## Max Rodgers - 27/04/2014 - Gerrard slips
ax.scatter("2014-04-27", data.loc[data['date_match'] == "2014-04-27", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
rodgersmax = ax.text(63, 1915, "Rodgers best\nElo", fontsize=13, ha="center", color="#2bb6e0", **hfont)
rodgersmax.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])


## Klopp first game - 17/10/2015
ax.scatter("2015-10-17", data.loc[data['date_match'] == "2015-10-17", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
klopp_firstgame = ax.text(105, 1710, "Klopp first\ngame", fontsize=13, ha="center", color="#2bb6e0", **hfont)
klopp_firstgame.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

## Coutinho Injury - 07/01/2016

## Man first game - 14/08/2016

## Salah first game - 12/08/2017
ax.scatter("2017-08-12", data.loc[data['date_match'] == "2017-08-12", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
salah_firstgame = ax.text(208, 1800, "Salah\narrival", fontsize=13, ha="center", color="#2bb6e0", **hfont)
salah_firstgame.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

## Van Dijk first game - 22/01/2018

## Alisson first game - 12/08/2018
ax.scatter("2018-08-12", data.loc[data['date_match'] == "2018-08-12", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
vandijk_firstgame = ax.text(253, 1878, "Alisson\narrival", fontsize=13, ha="center", color="#2bb6e0", **hfont)
vandijk_firstgame.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

## Champions League win - 01/06/2019
ax.scatter("2019-06-01", data.loc[data['date_match'] == "2019-06-01", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
ucl_win = ax.text(280, 2050, "Champions League\nWinner", fontsize=13, ha="center", color="#2bb6e0", **hfont)
ucl_win.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

## Last game before Corona - 07/03/2020
ax.scatter("2020-03-07", data.loc[data['date_match'] == "2020-03-07", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
corona = ax.text(350, 2050, "COVID", fontsize=13, ha="center", color="#2bb6e0", **hfont)
corona.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

## Title secured - 25/06/2020
ax.scatter("2020-06-24", data.loc[data['date_match'] == "2020-06-24", "Elo_Liverpool"], s=120, color="#2bb6e0", edgecolors="#08143D", lw=1, zorder=3 )
title_won = ax.text(357, 2018, "Title won", fontsize=13, ha="center", color="#2bb6e0", **hfont)
title_won.set_path_effects([path_effects.withStroke(linewidth=3, foreground="w")])

plt.show()
plt.tight_layout()
```
